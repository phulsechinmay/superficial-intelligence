```python
suggested = pd.read_csv("suggested_locations.csv", dtype={'id': 'category'})
comments = pd.read_csv("suggested_comments.csv", dtype={'id': 'category'})
```

```python
import requests
import urllib.parse
def request(text):
    print(text)
    try:
        annotations = requests.get(f"https://api.dandelion.eu/datatxt/nex/v1/?lang=en&text={urllib.parse.quote(text[:2500])}&include=image%2Calternate_labels%2Ctypes%2Cabstract%2Ccategories%2Clod&token=c9b309ecbe924a2ca626153d774de196").json()['annotations']
        print([a['title'] for a in annotations])
        return annotations
    except:
        print('error')
        return []
# request(comments.comment[comments.comment != 'I Agree'].comment.dropna().iloc[6])
isvalid = lambda c: isinstance(c, str) and c != 'I Agree'
comments['annotations'] = comments.apply(lambda r: request(r.comment) if isvalid(r.comment) else [], axis=1)
```
