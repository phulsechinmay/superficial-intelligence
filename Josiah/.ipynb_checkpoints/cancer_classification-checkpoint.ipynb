```python
import pandas as pd
import numpy as np
arr = np.array
import seaborn as sns
from math import ceil
from matplotlib import pyplot as plt
import datetime as dt
```

```python
dtype = {'Trip ID': np.str,
         'Starting Station ID': np.str,
         'Trip Route Category': 'category',
         'Passholder Type': 'category',
         'Ending Station ID': np.str,
         'Bike ID': np.str,
        }
df = pd.read_csv('data/metro-bike-share-trip-data.csv', dtype=dtype)
df = df.drop(['Starting Lat-Long', 'Ending Lat-Long'], axis=1)
df['Start Time'] = pd.to_datetime(df['Start Time'])
df['End Time'] = pd.to_datetime(df['End Time'])
df['Start Time (sec)'] = (df['Start Time'] - dt.datetime(1970,1,1)).dt.total_seconds()
df['End Time (sec)'] = (df['End Time'] - dt.datetime(1970,1,1)).dt.total_seconds()
df.head()
```

```python
# # what about outliers
# data = df['Duration']
# outliers = data[(np.abs(data-data.mean()) > (3*data.std()))]
# len(outliers)
# (data/3600).describe()
# 2% of our data is more than 10000 sec (2.78 hours)
# len(df['Duration'][df['Duration'] > 10000]) / len(df['Duration'])
df['Duration Trim'] = df['Duration'][df['Duration'] <= 10000]
```

```python
nrow = 2
ncol = 4
# figsize=(width,height) in inches
fig, axes = plt.subplots(nrow, ncol, figsize=(30, 10))
plt.subplots_adjust(hspace=.4, wspace=.4)

df.groupby('Passholder Type')['Passholder Type'].count().plot.bar(ax=axes[0,0])
df.groupby('Trip Route Category')['Trip Route Category'].count().plot.bar(ax=axes[0,1])
sns.distplot(df['Duration Trim'].dropna(), ax=axes[0,2])
plt.show()
```

```python
# Make n subplots for n features
# Each subplot should have the distribution for all classes graphed on it.
def make_displots(feature_names, data):
    nrow = ceil(len(feature_names) / 3)
    ncol = 4
    # figsize=(width,height) in inches
    fig, axes = plt.subplots(nrow, ncol, figsize=(10, 30))
    plt.subplots_adjust(hspace=.4, wspace=.4)
    # sns.set_style("darkgrid", {"axes.facecolor": ".9"})
    for i in range(nrow):
        for j in range(ncol):
            name = feature_names[(i+1)*(j+1)-1]
            sns.distplot(data[data.label == 'luminal A'][name], ax=axes[i,j])
            sns.distplot(data[data.label == 'luminal B'][name], ax=axes[i,j])
            sns.distplot(data[data.label == 'basal'][name], ax=axes[i,j])
            sns.distplot(data[data.label == 'Her2 positive'][name], ax=axes[i,j])
            axes[i,j].set_yticks([])
    fig.legend(['luminal A', 'luminal B', 'basal', 'Her2 positive'])
    plt.show()
```
